```python

import numpy as np 
import pandas as pd 
import matplotlib.pyplot as plt
import tensorflow as tf
import cv2
from PIL import Image
import os


data=[]
numbers=[]

height = 28
width = 28
channels = 3
classes = 43
n_inputs = height*width*channels

for i in range(classes) :
    path = "E:/proj/Train/{0}/".format(i)
    print(path)
    Class=os.listdir(path)
    end=0
    for a in Class:
        if end%2==0:
            image=cv2.imread(path+a)
            image_from_array = Image.fromarray(image, 'RGB')
            size_image = image_from_array.resize((height, width))
            data.append(np.array(size_image))
            numbers.append(i)
        end=end+1

            
Cells=np.array(data)
numbers=np.array(numbers)

s=np.arange(Cells.shape[0])
np.random.seed(43)
np.random.shuffle(s)
Cells=Cells[s]
numbers=numbers[s]

X_train=Cells
#(X_train,X_val)=Cells[(int)(0.3*len(numbers)):],Cells[:(int)(0.3*len(numbers))]
X_train = X_train.astype('float32')/255 
#X_val = X_val.astype('float32')/255
#(y_train,y_val)=numbers[(int)(0.3*len(numbers)):],numbers[:(int)(0.3*len(numbers))]
y_train=numbers

from keras.utils import to_categorical
y_train = to_categorical(y_train, 43)
#y_val = to_categorical(y_val, 43)


from keras.models import Sequential
from keras.layers import Conv2D, MaxPool2D, Dense, Flatten, Dropout



model = Sequential()
model.add(Conv2D(filters=32, kernel_size=(5,5), activation='relu', input_shape=X_train.shape[1:],name='a'))
model.add(Conv2D(filters=32, kernel_size=(5,5), activation='relu',name='b'))
model.add(MaxPool2D(pool_size=(2, 2),name='c'))
model.add(Dropout(rate=0.25))
model.add(Conv2D(filters=64, kernel_size=(3, 3), activation='relu',name='d'))
model.add(Conv2D(filters=64, kernel_size=(3, 3), activation='relu',name='e'))
model.add(MaxPool2D(pool_size=(2, 2),name='f'))
model.add(Dropout(rate=0.25))
model.add(Flatten())
model.add(Dense(256, activation='relu',name='g'))
model.add(Dropout(rate=0.5))
model.add(Dense(43, activation='softmax',name='preds'))

model.compile(
    loss='categorical_crossentropy', 
    optimizer='adam', 
    metrics=['accuracy']
)
from keras.preprocessing.image import ImageDataGenerator

datagen = ImageDataGenerator(
    #featurewise_center=True,
    #featurewise_std_normalization=True,
    rotation_range=7,
    width_shift_range=0.2,
    shear_range = 0.2,
    zoom_range = 0.2,
    height_shift_range=0.2
    #horizontal_flip=True
)
datagen.fit(X_train)
#history = model.fit_generator(datagen.flow(X_train, y_train, batch_size=32),
                    #steps_per_epoch=len(X_train) / 32, epochs=5,validation_data=(X_val,y_val))
history = model.fit_generator(datagen.flow(X_train, y_train, batch_size=20),
                    steps_per_epoch=len(X_train) / 2, epochs=6)
#history = model.fit(X_train, y_train, batch_size=30, epochs=5,validation_data=(X_val, y_val))

plt.figure(1)
plt.plot(history.history['accuracy'], label='training accuracy')
#plt.plot(history.history['val_acc'], label='val accuracy')
plt.title('Accuracy')
plt.xlabel('epochs')
plt.ylabel('accuracy')
plt.legend()

plt.figure(2)
plt.plot(history.history['loss'], label='training loss')
#plt.plot(history.history['val_loss'], label='val loss')
plt.title('Loss')
plt.xlabel('epochs')
plt.ylabel('loss')
plt.legend()

y_test=pd.read_csv("E:/proj/Test.csv")
numbers=y_test['Path'].values
y_test=y_test['ClassId'].values

data=[]

for n in numbers:
    image=cv2.imread('E:/proj/'+n)
    image_array = Image.fromarray(image, 'RGB')
    size_image = image_array.resize((height, width))
    data.append(np.array(size_image))

X_test=np.array(data)
X_test = X_test.astype('float32')/255 
pred = model.predict_classes(X_test)

from sklearn.metrics import accuracy_score
accuracy_score(y_test, pred)

import itertools
from sklearn.metrics import confusion_matrix

def plot_confusion_matrix(cm, classes,
                          normalize=False,
                          title='Confusion matrix',
                          cmap=plt.cm.Blues):
    plt.imshow(cm, interpolation='nearest', cmap=cmap)
    plt.title(title)
    plt.colorbar()
    tick_marks = np.arange(len(classes))
    plt.xticks(tick_marks, classes, rotation=75) 
    plt.yticks(tick_marks, classes)

    if normalize:
        cm = cm.astype('float') / cm.sum(axis=1)[:, np.newaxis]
        
    thresh = cm.max() / 2.
    for i, j in itertools.product(range(cm.shape[0]), range(cm.shape[1])):
            plt.text(j, i, cm[i, j],
            horizontalalignment="center",
            color="white" if cm[i, j] > thresh else "black")

    plt.tight_layout()
    plt.ylabel('True label')
    plt.xlabel('Predicted label')
    
class_names = range(43)
cm = confusion_matrix(pred,y_test)

plt.figure(2)
plot_confusion_matrix(cm, classes=class_names, title='Confusion matrix')

```
